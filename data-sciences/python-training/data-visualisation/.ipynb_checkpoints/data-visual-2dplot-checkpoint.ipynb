```python
#import matplot libraries
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#website traffic data
#number of users / visitors for the website
web_customers = [123,456,234,232,134,567,768,235]
time_hrs = [3,4,5,6,7,8,9,10]
```

```python
#style of the plot
style.use('ggplot')
#plot site traffic data (x axis hrs and y axis of customers)
plt.plot(time_hrs, web_customers)
#title of plot
plt.title('Web traffic data')
#label for X axis
plt.xlabel('Time in hours')
#label for Y axis
plt.ylabel('Number of users')
plt.show()
```

```python
#style of the plot
style.use('ggplot')
#plot site traffic data (x axis hrs and y axis of customers) with different line style
plt.plot(time_hrs, web_customers, color = 'g', linestyle = '--', linewidth = 2.5)
#title of plot
plt.title('Web traffic data')
#label for X axis
plt.xlabel('Time in hours')
#label for Y axis
plt.ylabel('Number of users')
plt.show()
```

```python
#style of the plot
style.use('ggplot')
#plot site traffic data (x axis hrs and y axis of customers) with different line style
plt.plot(time_hrs, web_customers, color = 'g', linestyle = '--', linewidth = 2.5)
#setting plot axis ranges
plt.axis([2.5,10.5,50,790]) #ranges are specified within a list
#title of plot
plt.title('Web traffic data')
#label for X axis 
plt.xlabel('Time in hours')
#label for Y axis
plt.ylabel('Number of users')
plt.show()
```

```python
#style of the plot
style.use('ggplot')
#plot site traffic data (x axis hrs and y axis of customers)
#also setting alpha for transparency
plt.plot(time_hrs, web_customers, alpha = 0.4)
plt.axis([2.5,10.5,50,900]) #ranges are specified within a list
#title of plot
plt.title('Web traffic data')

#annotating the plot ; coordinates of text and arrow are xytext and xy respectively
plt.annotate('Max', ha='center', va='bottom', xytext = (8,760), xy=(9,760), arrowprops = {'facecolor': 'green'})

#label for X axis 
plt.xlabel('Time in hours')
#label for Y axis
plt.ylabel('Number of users')
plt.show()
```

## Multiple plots example

```python
#website traffic data
#number of users / visitors for the website
web_monday = [123,456,234,232,134,567,768,235]
web_tuesday = [113,675,672,872,105,223,167,782]
web_wednesday = [102,902,254,729,123,112,643,982]

time_hrs = [3,4,5,6,7,8,9,10]
```

```python
#style of the plot
style.use('ggplot')
#plot site traffic data (x axis hrs and y axis of customers)
#monday plot
plt.plot(time_hrs, web_monday, alpha = 0.7, color = 'r', label = 'monday', linewidth = 0.5)
#tuesday plot
plt.plot(time_hrs, web_tuesday, alpha = 0.7, color = 'b', label = 'tuesday', linewidth = 1.0)
#wednesday plot
plt.plot(time_hrs, web_wednesday, alpha = 0.7, color = 'g', label = 'wednesday', linewidth = 0.5)

plt.axis([2.5,10.5,50,1100]) #ranges are specified within a list
#title of plot
plt.title('Web traffic data')

#label for X axis 
plt.xlabel('Time in hours')
#label for Y axis
plt.ylabel('Number of users')
plt.legend()
plt.show()
```

## Subplots

```python
#creating subplots side by side
plt.figure(figsize=(8,4))
plt.subplots_adjust(hspace=0.25)
plt.subplot(1,2,1)
plt.title('monday')
plt.plot(time_hrs, web_monday, color = 'r', linestyle = '-.', linewidth = 1)
plt.subplot(1,2,2)
plt.title('tuesday')
plt.plot(time_hrs, web_tuesday, color = 'g', linestyle = '--', linewidth = 0.7)
```
