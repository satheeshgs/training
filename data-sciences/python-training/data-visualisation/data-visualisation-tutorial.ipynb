```python
#import necessary libraries
import numpy as np
#import matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#generate random numbers
randomNumber = np.random.rand(20)
```

```python
#view the numbers
randomNumber
```

```python
#select style of the plot
style.use('ggplot')

#plot the random numbers
plt.plot(randomNumber, 'b', label='line one', linewidth = 2)

#label the axes
plt.xlabel('range')
plt.ylabel('numbers')

#title of plot
plt.title('first plt')
```

```python
plt.legend()
plt.show()
```
