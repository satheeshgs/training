```python
#import necessary libaries
from bs4 import BeautifulSoup
```

```python
#import the web scraping file

#change filepath as per need
#HTMLfilepath = "yahweh_wiki.hmtl"

with open("yahweh_wiki.html") as fp:
    soup = BeautifulSoup(fp, "lxml")
```

```python
#view content of soup object
soup.contents
```

```python
#search using find methods
tag_li = soup.find("li")
```

```python
#print the tag type
type(tag_li)
```

```python
#print the tag
tag_li
```

```python
#search the document using the find method for the ID
find_id = soup.find(title='Asherah')
```

```python
#print the find_id object
find_id
```

```python
#print the string value
find_id.string
```

```python
#search using string only
find_string = soup.find("Amun")
```

```python
#search using findall for specific class
find_class = soup.find_all(class_="mw-redirect")
```

```python
type(find_class)
```

```python
#printing first element of the resultant set
print(find_class[0])
```

```python
#printing second element of the resultant set
print(find_class[1])
```

```python
#finding parents of a specific class
find_class0 = find_class[0]
find_parent0 = find_class0.find_parents("ul")
```

```python
find_parent0
```

```python
next_sibling0 = find_class0.findNextSibling()
```

```python
next_sibling0
```

```python
previous_sibling0 = find_class0.find_previous_sibling()
```

```python
#use regular expressions to search the document 
import re
exmaple_ref = """
<a href=
"https://en.wikipedia.org/wiki/Abzu" title="Abzu"
>
Abzu/Apsu
</a>
"""
soup_ref = BeautifulSoup(exmaple_ref, 'lxml')
#use compile method to compile the information which contains regular expression
web_link_regexp = re.compile("https://en.wikipedia.org/wiki/\w+")
#find and print something using the regular expression
web_links = soup_ref.find(text=web_link_regexp)
```

```python
print(web_links)
```
